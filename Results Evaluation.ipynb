Evaluation metrics include: 

* `ave_ymin`: the average of minimal response values found over all replicates

* `ymin_achieve`: the best minimal response found

* `ymin_over_time`: the average of minimal response found over all replicates along time

* `ave_regrets`: the average of regrets 

* `ave_num_failures`: the average number of failures found

* `num_fail_rois`: the average number of ROIs

* `ratio_rois`: the ratio of detected all the ROIs

* `detect_rois_time`: the 1-th time detect the i-th ROIs for each replicate

* `n_founds_mat`: the number of times finding the i-th ROIs

* `n_roi_found`: the number of ROIs found for each replicate

* `area_founds`: the volume of convex hull(dim > 1) or interval(dim = 1) generated by points in ROIs   

* `detect_time_all`: the detection time for each replicate

* `failure_detected`: the locations of failure detected in each replicate

* `n_roi_over_time`: the number of ROIs detected along time for each replicate
